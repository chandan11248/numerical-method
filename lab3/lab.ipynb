## linear fitting 

```python
import numpy as np
import matplotlib.pyplot as plt


x = np.array([1, 2, 3, 4, 5])
y = np.array([2.2, 2.8, 3.6, 4.5, 5.1])


n = len(x)

sum_x = np.sum(x)
sum_y = np.sum(y)
sum_x2 = np.sum(x**2)
sum_xy = np.sum(x * y)

m = (n * sum_xy - sum_x * sum_y) / (n * sum_x2 - sum_x**2)
c = (sum_y - m * sum_x) / n

print(f"Slope (m): {m}")
print(f"Intercept (c): {c}")


y_pred = m * x + c


plt.figure(figsize=(8, 5))
plt.scatter(x, y, color='red', label='Data Points')
plt.plot(x, y_pred, color='blue', label='Best Fit Line')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Least Squares Linear Fit')
plt.legend()
plt.grid(True)
plt.show()
```

## exponential

## using crammers rule

```python
import numpy as np
import matplotlib.pyplot as plt


x = np.array([2, 4, 6, 8, 10,12])
y = np.array([3, 5, 10, 18, 48,90])


Y = np.log(y)


n = len(x)
Sx = np.sum(x)
SY = np.sum(Y)
Sx2 = np.sum(x**2)
SxY = np.sum(x * Y)

D = n * Sx2 - Sx * Sx


D_A = SY * Sx2 - Sx * SxY


D_b = n * SxY - SY * Sx

A = D_A / D
b = D_b / D
a = np.exp(A)

print(f"Exponential Model: y = {a:.4f} * e^({b:.4f}x)")


y_pred = a * np.exp(b * x)


plt.figure(figsize=(8, 5))
plt.scatter(x, y, color='red', label='Data Points')
plt.plot(x, y_pred, color='green', label='Exponential Fit (Cramer\'s Rule)')
plt.xlabel('x')
plt.ylabel('y')
plt.title(rf'Exponential Fit using Cramer\'s Rule: $y = {a:.2f} \cdot e^{{{b:.2f}x}}$')
plt.grid(True)
plt.legend()
plt.show()
```
